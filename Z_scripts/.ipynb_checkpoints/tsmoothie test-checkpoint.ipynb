```python
import numpy as np
import matplotlib.pyplot as plt
from tsmoothie.smoother import *
N=1000
window = 100
t = np.linspace(0,1*np.pi,N)
x = np.cos(t)
y = 2*np.sin(t) + np.random.random(N) * 0.2

# operate smoothing
smootherx = ConvolutionSmoother(window_len=5, window_type='ones')
smootherx.smooth(x)

# generate intervals
lowx, upx = smootherx.get_intervals('sigma_interval', n_sigma=2)

# operate smoothing
smoothery = ConvolutionSmoother(window_len=window, window_type='ones')
smoothery.smooth(y)

# generate intervals
lowy, upy = smoothery.get_intervals('sigma_interval', n_sigma=2)

# plot the smoothed timeseries with intervals
plt.figure(figsize=(11,6))
plt.plot(x, y, '.r')
plt.plot(smootherx.smooth_data[0], smoothery.smooth_data[0], linewidth=3, color='blue')

plt.fill(
    np.append(lowx, lowy),
    np.append(upx, upy),
)
# plt.fill_between(smootherx.data[0], smoothery.data[0], low[0], up[0], alpha=0.3)
# plt.fill_between(range(len(smoother.data[0])), low[0], up[0], alpha=0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tsmoothie.smoother import *
from np.random import permutation

x = 10+np.linspace(0,2*np.pi,100)
y = 1 + np.sin(x) + np.random.random(100) * 0.2
ind = permutation(len(x))
x = x[ind]
y = y[ind]

# operate smoothing
smoother = ConvolutionSmoother(window_len=5, window_type='ones')
smoother.smooth(y)

# generate intervals
low, up = smoother.get_intervals('sigma_interval', n_sigma=2)

# plot the smoothed timeseries with intervals
plt.figure(figsize=(11,6))
plt.plot(x, smoother.smooth_data[0], linewidth=3, color='blue')
plt.plot(x, smoother.data[0], '.k')
# plt.fill_between(smoother.data[0], low[0], up[0], alpha=0.3)
# plt.fill_between(smoother.data[0], low[0], up[0], alpha=0.3)
```
