```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')


def symbolic_fn(coef):
    """
    Generating function using coefficients
    The function takes the list of input coefficients and generate Symbolic function
    """
    sym_fun=0
    x=sym.Symbol('x')
    for i,val in enumerate(coef):
        sym_fun=sym_fun+coef[-1-i]*x**i
    return sym_fun


def evaluate_sym_exp(exp, value):
    """Evaluating the symbolic function"""
    return exp.subs(list(exp.free_symbols)[0], value)


def newtons_method(fn):
    """
    Newtons Method
    The function takes a symbolic expression as the input and gives one root of the function as output
    """
    n=0
    x=1
    while n < 10:
        fn_val=evaluate_sym_exp(fn, x)
        dif_fn_val=evaluate_sym_exp(sym.diff(fn), x)
        x = x - (fn_val/dif_fn_val)
        n = n + 1
    return round(x)


def synthetic_division(coef,root):
    """
    Synthetic Division
    It takes one of the roots of n-degree polynomial and outputs a polynomial of n-1 degree
    """
    quotient=[]
    val=coef[0]
    for i,j in enumerate(coef):
        if i==0:
            quotient.append(coef[i])
        else:
            val=val*root+coef[i]
            quotient.append(val)
    quotient.pop()
    return (quotient)


def solve_any_poly(coef):
    """
    Solving Polynomial
    The function takes the list of coefficients of the polynomial of any degree
    and outputs list of all roots of the given polynomial
    """
    roots=[]
    for i,j in enumerate(coef):
        while len(coef)>2:
            fn = symbolic_fn(coef)
            print(fn)
            root=newtons_method(fn)
            roots.append(root)
            return root # !!!!!
            coef=synthetic_division(coef,root)
    return roots + [-coef[-1]]


n = int(input('enter n:'))
# coeff = [1, -14, 33, 80, -100]
coeff_str = str(input('Enter Coefficients of Equation:'))
coeff = [ int(item) for item in coeff_str.split() ]
roots = solve_any_poly(coeff)
# plot_fn(coeff, roots)
print(roots)
```

```python
from scipy.optimize import fsolve

n = int(input())
# s = input()
s = '10000 1 -10000 -10000 2 4000'

coefs = tuple(map(float, s.split()))

def fun(x, coefs, n):
    res = 0
    for i in range(n+1):
        res += coefs[i]*x**i
    return res

def dfun(x, coefs, n):
    res = 0
    for i in range(1,n+1):
        res += i*coefs[i]*x**(i-1)
    return res
print("{:40}".format(fsolve(fun, [1], fprime=dfun, args=(coefs, n))[0]))
```

```python
def find_index(arr, K):
    n = len(arr)
    # Traverse the array
    for i in range(n):
        # If K is found
        if arr[i] == K:
            return i
        # If arr[i] less K
        elif arr[i] < K:
            return i 
    # If all array elements are smaller
    return n
# n = 4
n = int(input("Enter the list size, 2<=n<=10^5:"))
# numbers_list = [10, 11, 12, 13]
s = input("User list is: ")
numbers_list = tuple(map(float, s.split()))
numbers_list = sorted(numbers_list, reverse=True)
com = 0
while len(numbers_list)>1:
    sums = numbers_list[-1] + numbers_list[-2]
    com += sums*0.05
    numbers_list.pop()
    numbers_list.pop()
    
    pos = find_index(numbers_list, sums)
    numbers_list.insert(pos, sums)
print(numbers_list, com)
```
